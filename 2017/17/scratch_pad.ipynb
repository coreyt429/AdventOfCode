```python
# works for part 1, don't try it for part 2
steps=359
curr_pos = 0
spinlock = [0]
for idx in range(1, 2017 + 1):
    curr_pos = (curr_pos + steps) % len(spinlock)
    curr_pos += 1
    spinlock.insert(curr_pos, idx)

print(curr_pos, spinlock)
print(spinlock[curr_pos + 1])
```

```python
from collections import deque

steps = 359
spinlock = deque([0])

for i in range(1, 2017 + 1):
    spinlock.rotate(-steps)
    spinlock.append(i)

print(spinlock[0])
```

```python
from collections import deque

steps = 359
spinlock = deque([0])

for i in range(1, 50000001):
    spinlock.rotate(-steps)
    spinlock.append(i)

print(spinlock[spinlock.index(0) + 1])
```
