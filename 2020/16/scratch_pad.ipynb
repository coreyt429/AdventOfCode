```python
import os
import sys
sys.path.append(os.path.realpath('../..'))
import aoc
my_aoc = aoc.AdventOfCode(2020,16)
from solution import parse_input, parse_rules, rule_match, reduce_fields, identify_fields
```

```python
input_text = """class: 1-3 or 5-7
row: 6-11 or 33-44
seat: 13-40 or 45-50

your ticket:
7,1,14

nearby tickets:
7,3,47
40,4,50
55,2,20
38,6,12"""




rules, my_ticket, nearby_tickets = parse_input(input_text)
valid_nums = set()
for num_set in rules.values():
    valid_nums = valid_nums.union(num_set)

invalid_nums = []
valid_tickets = [my_ticket]
for ticket in nearby_tickets:
    valid = True
    for num in ticket:
        if num not in valid_nums:
            valid = False
            invalid_nums.append(num)
    if valid:
        valid_tickets.append(ticket)
    
print(f"Part 1: {sum(invalid_nums)}")
print(f"valid_tickets: {valid_tickets}")

fields = identify_fields(valid_tickets, rules)
print(f"reduced fields: {fields}")
    
```
