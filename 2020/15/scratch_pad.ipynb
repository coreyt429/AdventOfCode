```python
input_text="""0,3,6"""
numbers = [int(num) for num in input_text.split(',')]
print(numbers)
```

```python
from collections import deque
dq = deque(reversed(numbers))
turn = 1
for number in dq:
    print(f"turn: {turn}, spoken: {number}")
    turn += 1
for _ in range(turn, 2021):
    current = dq[0]
    if dq.count(current) == 1:
        dq.appendleft(0)
    else:
        dq.appendleft(dq.index(current, 1))
    print(f"turn: {turn}, spoken: {dq[0]}")
    turn += 1
print(dq)
```

```python
input_text="""0,3,6"""
numbers = [int(num) for num in input_text.split(',')]
print(numbers)
from collections import defaultdict
seen = defaultdict(list)
turn = 1
for current in numbers:
    # print(f"turn: {turn}, spoken: {current}")
    seen[current].append(turn)
    turn += 1
target = 2020 + 1
part = 2
if part == 2:
    target = 30000000 + 1
for _ in range(turn, target):
    # print(turn, current, seen[current])
    if len(seen[current]) == 1:
        seen[0].append(turn)
        current = 0
    else:
        current = seen[current][-1] - seen[current][-2]
        seen[current].append(turn)
    if turn % 1000 == 0:
        print(f"turn: {turn}, spoken: {current}")
    turn += 1
turn -= 1
print(f"turn: {turn}, spoken: {current}")
```
