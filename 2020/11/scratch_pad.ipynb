```python
import os
import sys
sys.path.append(os.path.realpath('../..'))
import aoc
my_aoc = aoc.AdventOfCode(2020,11)
from grid import Grid
```

```python
input_text = """L.LL.LL.LL
LLLLLLL.LL
L.L.L..L..
LLLL.LL.LL
L.LL.LL.LL
L.LLLLL.LL
..L.L.....
LLLLLLLLLL
L.LLLLLL.L
L.LLLLL.LL"""
```

```python
waiting_area = Grid(input_text, use_overrides=False)
last = str(waiting_area)
while True:
    tmp_dict = {}
    for seat in waiting_area:
        neighbors = waiting_area.get_neighbors(point=seat, directions='nesenwsw')
        neighbor_string = ''
        for neighbor in neighbors.values():
            neighbor_string += waiting_area.get_point(neighbor)
        # If a seat is empty (L) and there are no occupied seats adjacent to it, the seat becomes occupied.
        if waiting_area.get_point(seat) =='L' and neighbor_string.count('#') == 0:
            tmp_dict[seat] = '#'
        # If a seat is occupied (#) and four or more seats adjacent to it are also occupied, the seat becomes empty.
        if waiting_area.get_point(seat) =='#' and neighbor_string.count('#') >= 4:
            tmp_dict[seat] = 'L'
        # Otherwise, the seat's state does not change.
    for seat in tmp_dict:
        waiting_area.set_point(seat, tmp_dict[seat])
    if last == str(waiting_area):
        break
    last = str(waiting_area)
print(str(waiting_area).count('#'))
 
```

```python
waiting_area = Grid(input_text, use_overrides=False, default_value='x', ob_default_value='X')

last = str(waiting_area)
print(f"{waiting_area}\n")

def next_seat(waiting_area, seat, direction):
    neighbors = waiting_area.get_neighbors(point=seat, directions=[direction])
    if not neighbors:
        return '.'
    state = waiting_area.get_point(neighbors[direction])
    if state in 'L#':
        return state
    return next_seat(waiting_area, neighbors[direction], direction)

while True:
    tmp_dict = {}
    for seat in waiting_area:
        neighbor_string = ''
        for direction in ['n','s','e','w','ne','nw','se','sw']:
            neighbor_string += next_seat(waiting_area, seat, direction)
        # If a seat is empty (L) and there are no occupied seats adjacent to it, the seat becomes occupied.
        if waiting_area.get_point(seat) =='L' and neighbor_string.count('#') == 0:
            tmp_dict[seat] = '#'
            continue
        # If a seat is occupied (#) and four or more seats adjacent to it are also occupied, the seat becomes empty.
        if waiting_area.get_point(seat) =='#' and neighbor_string.count('#') >= 5:
            tmp_dict[seat] = 'L'
            continue    
        # Otherwise, the seat's state does not change.
    for seat in tmp_dict:
        waiting_area.set_point(seat, tmp_dict[seat])
    print(f"{waiting_area}\n")
    if last == str(waiting_area):
        break
    last = str(waiting_area)
print(str(waiting_area).count('#'))
```
