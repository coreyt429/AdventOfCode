```python

def test_loop(program):
    hgc = GameConsole(program)
    seen = set()
    while 0 <= hgc.ptr < len(hgc.program):
        print(f"{hgc.ptr} {hgc.accumulator}")
        if hgc.ptr in seen:
            return False
        seen.add(hgc.ptr)
        hgc.step()
    print("Program exited")
    return hgc.accumulator

init_program = input_text.splitlines()
for idx, line in enumerate(init_program):
    # No acc instructions were harmed in the corruption of this boot code.
    if 'acc' in line:
        continue
    test_program = input_text.splitlines()
    print(f"before: {test_program[idx]}")
    if "nop" in line:
        test_program[idx] = test_program[idx].replace('nop', 'jmp')
    else:
        test_program[idx] = test_program[idx].replace('jmp', 'nop')
    print(f"after: {test_program[idx]}")
    print(f"{idx}: {line} {result}")
    result = test_loop(test_program)
    if result:
        print(result)
        break


```
