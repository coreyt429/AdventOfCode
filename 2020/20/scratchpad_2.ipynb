```python
def status(tile):
    print()
    print(tile)
    print("Mates:")
    for mate_id in tile.mates:
        print(f"  {tile.parent.tile_by_id(mate_id)}")
    print(tile.str_grid())
    print()


live_data = ''
with open('input.txt') as file:
    live_data = file.read().rstrip()

tile_collection = Tiles(live_data)
tile_collection.refresh()
corners = []
for tile in tile_collection:
    # print(f"{tile} - {len(tile.mates)} {tile.mates}")
    if tile.variety == 'corner':
        corners.append(tile.tile_id)
tile = tile_collection.tile_by_id(corners[0])
tile.swap(0)
status(tile)

for position in (range(len(tile_collection))):
# for position in (0, 1, 2, 3, 4, 12, 13, 14, 15, 16):
    print(f"Position: {position}")
    tile_collection.refresh()
    tile = tile_collection.tile_by_position(position)
    tile.orient_self()
    tile.orient_mates()

print(tile_collection.str_grid())


```
