```python
import re

def parse_data(lines):
    passwords = []
    pattern_rule = re.compile(r'(\d+)-(\d+) (\w): (\S+)')
    for line in lines:
        match = pattern_rule.match(line)
        if match:
            passwords.append(
                {
                    "password": match.group(4),
                    "min": int(match.group(1)),
                    "max": int(match.group(2)),
                    "letter": match.group(3)
                }
            )
    return passwords

def is_valid(pass_data):
    count = pass_data['password'].count(pass_data["letter"])
    return pass_data['min'] <= count <= pass_data['max']
```

```python
input_text = """1-3 a: abcde
1-3 b: cdefg
2-9 c: ccccccccc"""

data = parse_data(input_text.splitlines())
for password in data:
    print(password, is_valid(password))
```
