```python
import itertools
point = (0, 0)

from itertools import product

# @lru_cache(maxsize=None)
def get_offsets(size):
    """
    Function to get the neighbor offsets for a point dimension size
    """
    return tuple(product([-1, 0, 1], repeat=size))

# @lru_cache(maxsize=None)
def get_neighbors_itertools(point):
    """
    Function to calculate neighbors of a point of arbitrary dimension
    """
    # print(f"get_neighbors_itertools({point})")
    # Generate all offsets (-1, 0, 1) for each element in the tuple
    offsets = get_offsets(len(point))
    # Add each offset to the original tuple to get the neighboring tuples
    neighbors = []
    for offset in offsets:
        # print(offset)
        neighbor = tuple((value + offset[idx] for idx, value in enumerate(point)))
        print(neighbor)
        # for idx, value in enumerate(point):
        #     neighbor.append(value + offset[idx])
        neighbors.append(neighbor)
    # print(neighbors)
    # Convert generator expressions into lists of tuples
    return [tuple(neighbor) for neighbor in neighbors if neighbor != list(point)]
```
