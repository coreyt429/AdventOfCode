```python
input_text = """35
20
15
25
47
40
62
55
65
95
102
117
150
182
127
219
299
277
309
576"""
```

```python
from itertools import combinations
preamble_size=5
numbers = [int(num) for num in input_text.splitlines()]
print(numbers)
for idx in range(preamble_size, len(numbers) + 1):
    print(f"{idx}: {numbers[idx]} {numbers[idx-preamble_size:idx]}")
    match_found = False
    for combo in combinations(numbers[idx-preamble_size:idx], 2):
        if sum(combo) == numbers[idx]:
            match_found = True
            break # inner for loop
    if match_found:
        continue
    print(numbers[idx])
    break # outer for loop
```
