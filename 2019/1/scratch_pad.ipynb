```python
def required_fuel(mass):
    """part 1 fuel calculation"""
    return mass // 3 -2

def required_fuel_2(mass):
    """part 2 recursive fuel calculation"""
    fuel = mass // 3 -2
    if fuel < 0:
        return 0
    return fuel + required_fuel_2(fuel)
```

```python
for m in [12, 14, 1969, 100756]:
    print(f"{m}: {required_fuel(m)}")
```

```python
for m in [12, 14, 1969, 100756]:
    print(f"{m}: {required_fuel_2(m)}")
```
