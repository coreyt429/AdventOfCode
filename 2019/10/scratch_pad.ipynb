```python
def get_sorted_points(line_list, laser):
    lines = []
    for line in line_list:
        lines.append(sort_collinear_points(line + [laser])) 
    # init new_lines
    new_lines = []
    # iterate over lines
    for line in lines:
        # if laser is on the front end, just remove it and leave line alone
        if line[0] == laser:
            line.pop(0)
            new_lines.append(line)
            continue
        # if laser is on the back end, remove it, and reverse the line
        if line[-1] == laser:
            line.pop(-1)
            new_lines.append(line[::-1])
            continue
        print(line)
        # Find the index of the laser point in the line
        index = line.index(laser)
        # reverse the line leading to laser
        new_lines.append(line[:index][::-1])
        # add the line after laser
        new_lines.append(line[index + 1:])

    new_lines.sort(key=lambda line: bearing(laser, line[0]))
    sorted_points = []
    added = True
    while added:
        added = False
        for line in new_lines:
            if not line:
                continue
            sorted_points.append(line.pop(0))
            added = True
    return sorted_points


for test_case in [test_case_large]:
    grid = Grid(test_case, use_overrides=False)
    for point in grid:
        if grid.get_point(point) == 'X':
            grid.set_point(point, '#')
            break
    laser, count, lines_by_point = find_best_position(grid)
    grid.set_point(laser, 'X')
    sorted_points = get_sorted_points(lines_by_point[laser], laser)
    lines = []
    
    for idx, point in enumerate(sorted_points):
        if idx == 199:
            print(f"The {idx + 1}th asteroid to be vaporized is at {point[0]},{point[1]}.")



```
