```python
test_cases = [
    {
        "input_signal": '12345678',
        "phases": 4,
        "final_result": '01029498'
    },
    {
        "input_signal": '80871224585914546619083218645595',
        "phases": 100,
        "final_result": '24176176'
    },
    {
        "input_signal": '19617804207202209144916044189917',
        "phases": 100,
        "final_result": '73745418'
    },
    {
        "input_signal": '69317163492948606335995924319873',
        "phases": 100,
        "final_result": '52432133'
    }
]

for idx, test_case in enumerate(test_cases):
    signal = [int(char) for char in test_case["input_signal"]]
    # print(f"\Input signal: {signal_to_str(signal)}\n")
    for _ in range(1, test_case["phases"] + 1):
        signal = run_phase(signal)
        # print(f"\nAfter {phase} phases: {signal_to_str(signal)}\n")
    result = signal_to_str(signal)
    # print(f"{result[:8]} == {test_case['final_result']}")
    assert result[:8] == test_case["final_result"]
    print(f"test_case {idx} passed: {result[:8]}")
```
