```python

counts = {0: 10, 1: 100}
for idx, test_case in enumerate(test_cases):
    moons = parse_input(test_case.splitlines())
    for tick in range(counts[idx]):
        moons = move_moons(moons)
    last_energy = calculate_energy(moons)
    print(last_energy)
```

```python
import math
from functools import reduce

X=0
Y=1
Z=2

def lcm_of_list(numbers):
    return reduce(math.lcm, numbers)

def moon_state(moon):
    return tuple([moon["id"],tuple(moon["position"]),tuple(moon["velocity"])])

def plane_state(moons, axis):
    state = [axis]
    for moon in moons:
        state.append(moon["position"][axis])
        state.append(moon["velocity"][axis])
    return tuple(state)
        

test_case = test_cases[1]
moons = parse_input(test_case.splitlines())
seen = set()
cycles = {}
last = {}
counter = 0
keep_going = True
periods = {}
while keep_going:
    for axis in [X, Y, Z]:
        state = plane_state(moons, axis)
        if state in seen and axis not in periods:
            periods[axis] = counter
            # print(counter, periods)
            if len(periods) == 3:
                keep_going = False
        else:
            seen.add(state)
    moons = move_moons(moons)
    counter += 1
print(lcm_of_list(periods.values()))


# We are on the right track here, but I think we need to track, X, Y, and Z seperately, to determine the repeating cycles


```
