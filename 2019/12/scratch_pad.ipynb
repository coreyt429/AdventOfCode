```python
import os
import sys
sys.path.append(os.path.realpath('../..'))
import aoc
my_aoc = aoc.AdventOfCode(2019,12)
from solution import parse_input, calculate_energy, lcm_of_list, move_moons, plane_state, moon_state
```

```python
test_cases = ["""<x=-1, y=0, z=2>
<x=2, y=-10, z=-7>
<x=4, y=-8, z=8>
<x=3, y=5, z=-1>""",
"""<x=-8, y=-10, z=0>
<x=5, y=5, z=10>
<x=2, y=-7, z=3>
<x=9, y=-8, z=-3>"""
]
```

```python

counts = {0: 10, 1: 100}
for idx, test_case in enumerate(test_cases):
    moons = parse_input(test_case.splitlines())
    for tick in range(counts[idx]):
        moons = move_moons(moons)
    last_energy = calculate_energy(moons)
    print(last_energy)
```

```python
X=0
Y=1
Z=2

for test_case in test_cases:
    moons = parse_input(test_case.splitlines())
    seen = set()
    cycles = {}
    last = {}
    counter = 0
    keep_going = True
    periods = {}
    while keep_going:
        for axis in [X, Y, Z]:
            state = plane_state(moons, axis)
            if state in seen and axis not in periods:
                periods[axis] = counter
                # print(counter, periods)
                if len(periods) == 3:
                    keep_going = False
            else:
                seen.add(state)
        moons = move_moons(moons)
        counter += 1
    print(lcm_of_list(periods.values()))
```
