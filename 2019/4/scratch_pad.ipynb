```python
import os
import sys
sys.path.append(os.path.realpath('../..'))
import aoc
my_aoc = aoc.AdventOfCode(2019,4)
```

```python
input_text = "123257-647015"
start_num, end_num = [int(num) for num in input_text.split('-')]
```

```python
def check_passcode(int_in, mode=1):
    work_str = str(int_in)
    has_repeating_digit = False
    no_incrementing_values = True
    length = len(work_str)
    for idx, char in enumerate(work_str):
        if idx < length - 1:
            if work_str[idx + 1] < char:
                no_incrementing_values = False
                break
            if char == work_str[idx + 1]:
                if mode == 2:
                    try:
                        if char == work_str[idx + 2]:
                            continue
                    except IndexError:
                        pass
                    try:
                        if char == work_str[idx - 1]:
                            continue
                    except IndexError:
                        pass
                has_repeating_digit = True
    return all([has_repeating_digit, no_incrementing_values])    

num = start_num
passwords = []
for num in range(start_num, end_num + 1):
    if check_passcode(num):
        passwords.append(num)

print(len(passwords))
```
