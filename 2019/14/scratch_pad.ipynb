```python
for idx, test_case in enumerate(test_cases[1:]):
    target = 1000000000000
    nano_factory  = NanoFactory(test_case)
    nano_factory.make(1, 'FUEL')
    current_count = target // nano_factory.ore_collected
    last_count = current_count
    last_ore = nano_factory.ore_collected
    step_size = 10000000
    while last_ore <= target or step_size >= 2:
        if last_ore > target:
            step_size = step_size // 2
            current_count = last_count + step_size
        else:
            last_count = current_count
            current_count += step_size
        # print(f"Trying: {current_count}")
        nano_factory  = NanoFactory(test_case)
        nano_factory.make(current_count, 'FUEL')
        # print(current_count, nano_factory.ore_collected)
        last_ore = nano_factory.ore_collected
    # print(last_count)
    
    print(idx, last_count, nano_factory.ore_collected, nano_factory.production['FUEL'], nano_factory.inventory['ORE'])
```
