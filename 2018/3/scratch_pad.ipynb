```python
input_text="""#1 @ 1,3: 4x4
#2 @ 3,1: 4x4
#3 @ 5,5: 2x2"""
input_lines = input_text.splitlines()
```

```python
import re

pattern_input = re.compile(r'#(\d+)\s+@\s+(\d+),(\d+): (\d+)x(\d+)')

def parse_input(lines):
    squares = []
    for line in lines:
        match = pattern_input.match(line)
        if match:
            squares.append({
               "id": int(match.group(1)),
               "col": int(match.group(2)),
               "row": int(match.group(3)),
               "width": int(match.group(4)),
               "height": int(match.group(5)),
            })
    return squares

def find_overlaps(squares):
    occupied = []
    for square in squares:
        for col in range(square['col'], square['col'] + square['width']):
            for row in range(square['row'], square['row'] + square['height']):
                occupied.append(tuple([row, col]))
    unique = set(occupied)
    duplicates = set()
    for square in unique:
        if occupied.count(square) > 1:
            duplicates.add(square)
    return len(duplicates)


squares = parse_input(input_lines)
overlap = find_overlaps(squares)
print(overlap)
```

```python
def squares_overlap(sq_1, sq_2):
    #sq_2: [(13, 146), (13, 170), (37, 170), (37, 146)]
    min_row = min([point[0] for point in sq_2])
    max_row = max([point[0] for point in sq_2])
    min_col = min([point[1] for point in sq_2])
    max_col = max([point[1] for point in sq_2])
    sq_overlap= False
    for point in sq_1:
        row, col = point
        if min_row <= row <= max_row and min_col <= col <= max_col:
            sq_overlap = True
    return sq_overlap
```

```python
sq_1 = [(272, 511), (272, 520), (284, 520), (284, 511)]
sq_2 = [(0, 0), (0, 1000), (1000, 1000), (1000, 0)]
squares_overlap(sq_1, sq_2)
```
