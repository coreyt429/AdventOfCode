```python
def check_sum(box_ids):
    matches = {
        2: set(),
        3: set()
    }
    for box_id in box_ids:
        for char in set(list(box_id)):
            count = box_id.count(char)
            if count in matches:
                matches[count].add(box_id)
    print(matches)
    return len(matches[2]) *  len(matches[3])

check_sum(input_lines)
    
```

```python
input_text = """abcde
fghij
klmno
pqrst
fguij
axcye
wvxyz"""
input_lines = input_text.splitlines()

def find_matches(box_ids):
    potentials = set()
    for box_id in box_ids:
        mid = len(box_id) // 2
        box_id_1 = ([box_id[:mid], box_id[mid:]])
        for box_id2 in box_ids:
            if box_id == box_id2:
                continue
            box_id_2 = ([box_id2[:mid], box_id2[mid:]])
            print(f"{box_id_1[0]} == {box_id_2[0]} or {box_id_1[1]} == {box_id_2[1]}: {box_id_1[0] == box_id2[0]} or {box_id_1[1] == box_id_2[1]}")
            if box_id_1[0] == box_id_2[0] or box_id_1[1] == box_id_2[1]:
                potentials.add(tuple(sorted([box_id, box_id2])))
    print(potentials)


find_matches(input_lines)
```
