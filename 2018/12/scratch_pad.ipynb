```python
input_text = """initial state: #..#.#..##......###...###

...## => #
..#.. => #
.#... => #
.#.#. => #
.#.## => #
.##.. => #
.#### => #
#.#.# => #
#.### => #
##.#. => #
##.## => #
###.. => #
###.# => #
####. => #"""
```

```python
def parse_input(input_text):
    (init_state, line_text) = input_text.split("\n\n")
    state = {}
    for idx, pot in enumerate(init_state.split(' ')[2]):
        if pot == '#':
            state[idx] = pot
    rules = {}
    for line in line_text.splitlines():
        rule, _, value = line.split(' ')
        rules[rule] = value
    return state, rules

def print_generations(generations):
    for idx, generation in enumerate(generations):
        print(f"{idx:2}:  ", end="")
        for pot in range(-3,37):
            print(f"{generation.get(pot,'.')} ",end="")
        print()

def next_generation(generation, rules):
    min_key = min(generation.keys())
    max_key = max(generation.keys)
    new_generation = {}
    for idx in range(min_key, max_key + 1):
        selection = ''.join(generation.get(pot, '.' ) for pot in range(idx - 1, idx + 2))
        value = rules.get(selection,'.')
        if value == '#':
            new_generation[idx] = value
    return new_generation
        


```

```python
init_state, spread_rules = parse_input(input_text)
generations = []
generations.append(init_state)
generations.append(next_generation(generations[-1], spread_rules))
print_generations(generations)
```
