```python
def generate_pairs():
    new_pairs = []
    for idx in range(65, 91):
        new_pairs.append(f"{chr(idx)}{chr(idx+32)}")
        new_pairs.append(f"{chr(idx+32)}{chr(idx)}")
    return new_pairs

# this was a slower method, when I thought my first method didn't work
# my original method is faster, and in the solution
# If I haven't said it before, READ THE INSTRUCTIONS, don't be like
# me sending the wrong data in :)
def reduce_string(pairs, start_string):
    print(pairs)
    working_list = list(start_string)
    idx = 0
    while idx < len(working_list) -1 :
        if f"{working_list[idx]}{working_list[idx + 1]}" in pairs:
            working_list.pop(idx + 1)
            working_list.pop(idx)
            idx -= 1
            continue
        idx += 1
    return ''.join(working_list)


pairs = generate_pairs()
reduced_string = reduce_string(pairs, "dabAcCaCBAcCcaDA") 
print(f"reduced_string: [{reduced_string}]")
```
