```python
input_text="""#########
#G..G..G#
#.......#
#.......#
#G..E..G#
#.......#
#.......#
#G..G..G#
#########"""
input_lines  = input_text.splitlines()
```

```python
class Player:
    def __init__(self, grid, team, pos, attack=3, hit_points=200):
        self.grid = grid
        self.team = team
        self.pos = pos
        self.attack = attack
        self.hit_points = hit_points
        self.grid.map[pos] = '.'
        self.grid.overrides[pos] = self.team

players = []
my_map = Grid(input_lines)
last_y = 0
for point in my_map:
    team = my_map.get_point(point)
    if team in ['G','E']:
        players.append(Player(my_map, team, point))
print(my_map)
print()
```
