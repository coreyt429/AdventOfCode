```python
for input_text in games:
    input_lines = input_text.splitlines()
    players = Players()
    my_map = Grid(input_lines)
    for point in my_map:
        team = my_map.get_point(point)
        if team in ['G','E']:
            players.append(Player(players, my_map, team, point))
    #print(my_map)
    completed_turns = 0
    game_over = False
    while not game_over:
        players.sort()
        for player in players:
            if not player.play_turn():
                game_over = True
                break
        if game_over:
                break
        completed_turns += 1
        #for player in players:
        #    if player.hit_points <= 0:
        #        player.death()
        #        continue
           
            
    
    remaining_hit_points = [player.hit_points for player in players]
    result = completed_turns * sum(remaining_hit_points)
    print(f"Outcome: {completed_turns} * {sum(remaining_hit_points)} = {result}")
```

```python
my_map = Grid(input_lines)
players = Players()
my_map = Grid(input_lines)
debug = False
for point in my_map:
    team = my_map.get_point(point)
    if team in ['G','E']:
        players.append(Player(players, my_map, team, point))
for turn in range(1,2):
    for player in players:
        player.play_turn()
    print(turn)
    for idx, line in enumerate(str(my_map).splitlines()):
        print(f"{idx:2} {line}", end="")
        for player in players:
            if player.pos[1] == idx:
                print(f" {player.hit_points}", end="")
        print()
for idx in range(20):
   if players.players[idx].pos == (11, 9):
       player = players.players[idx]
       break
   players.players[idx].play_turn
print(f"player: {player}")
print(players)
for idx, line in enumerate(str(my_map).splitlines()):
    print(f"{idx:2} {line}", end="")
    for playa in players:
        if playa.pos[1] == idx:
            print(f" {playa.hit_points}", end="")
    print()
opponents = player.find_opponents()
# for opponent in opponents:
#      print(f"opponent: {opponent}")
can_attack, targets = player.can_attack()
player.build_graph()
closest_distance = float('infinity')
shortest_paths = []
for open_square in player.find_open_squares(opponents):
    print(f"Trying: {open_square}")
    try:
        path = networkx.shortest_path(player.graph, source=player.pos, target=open_square, weight='weight')
        print(f"path: {path}")
    except networkx.exception.NetworkXNoPath:
        print(f"No Path")
        continue
    print(f"{len(path)} < {closest_distance}")
    if len(path) < closest_distance:
        closest_distance = len(path)
        shortest_paths = [path]
    elif len(path) == closest_distance:
        shortest_paths.append(path)
for path in shortest_paths:
    print(f"shortest_paths: {path}")

nearest_targets = [path[-1] for path in shortest_paths]
print(f"nearest_targets: {nearest_targets}")
#if debug: print(f"nearest_targets: {nearest_targets}")
# choose the target that comes first in the original order
opponent_set = set()
opponent_dict = {}
for target in nearest_targets:
    # get the opponent(s) square can attack
    _, opponents = player.square_can_attack(target)
    opponent_dict[target] = opponents
    opponent_set.update([opponent.pos for opponent in opponents])
print(f"opponent_set: {opponent_set}")
# get the closest opponent
nearest_opponent = player.eval_points(opponent_set)
print(f"nearest_opponent: {nearest_opponent}")
possible_targets = set()
for target in nearest_targets:
    for opponent in opponent_dict[target]:
        if opponent.pos == nearest_opponent:
            possible_targets.add(target)
# which target matches our selection rule
print(f"possible_targets: {possible_targets}")
target = player.eval_points(possible_targets)
print(f"target: {target}")
target = player.eval_points(possible_targets)
#if debug: print(f"target: {target}")
player.graph.add_node(target)
neighbors = player.grid.get_neighbors(point=target, directions=player.directions, invalid=['#', 'G', 'E'])
# invert weights, since direction is releative to target
weights = {
    's': .01,
    'e': .02,
    'w': .03,
    'n': .04
}
for direction, neighbor in neighbors.items():
    player.graph.add_edge(neighbor, target, weight=1)
try:
    path = networkx.shortest_path(player.graph, source=player.pos, target=target, weight='weight')
except networkx.exception.NetworkXNoPath:
    print(f"No path from {player.pos} to {target}")
    player.graph.remove_node(target)
player.graph.remove_node(target)
print(f"path: {path}")
player.move(path[1])
print(player)
```
