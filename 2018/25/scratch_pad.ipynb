```python
from heapq import heappop, heappush

def parse_data(lines):
    points = []
    for line in lines:
        data = [int(num) for num in line.split(',')]
        points.append(tuple(data))
    return points

def count_constellations_heap(points):
    origin = points[0]
    heap = []
    for point in points:
        heappush(heap, (0, manhattan_distance(origin, point), point))

    constellations = []
    last_length = 0
    counter = 0
    while heap:
        counter += 1
        if counter > 1000:
            print("Breaker")
            break

        attempt, distance, point = heappop(heap)
        placed = False
        # first constellation
        if len(constellations) == 0:
            constellations.append([point])
            continue

        for constellation in constellations:
            for point_b in constellation:
                if manhattan_distance(point, point_b) <= 3:
                    constellation.append(point)
                    placed = True
                    break
        
        if not placed:
            if attempt  > len(points):
                constellations.append([point])
            else:
                heappush(heap, (attempt + 1, manhattan_distance(origin, point), point))
    
    return len(constellations)

def combine_constellations(constellations):
    changed = True
    while changed:
        changed = False
        removes = []
        for con_a in constellations:
            if changed:
                continue
            list_a = list(con_a)
            for con_b in constellations:
                if con_b == con_a:
                    continue
                if changed:
                    continue
                list_b = list(con_b)
                for point_a in list_a:
                    if changed:
                        continue
                    for point_b in list_b:
                        if changed:
                            continue
                        print(point_a, point_b, manhattan_distance(point_a, point_b))
                        if manhattan_distance(point_a, point_b) <= 3:
                            con_a.update(con_b)
                            removes.append(con_b)
                            changed = True
        for con in removes:
            constellations.pop(constellations.index(con))
        
        print("Constellations:")
        for constellation in constellations:
            print(constellation)
        return constellations


from collections import deque

def group_points(points, distance_threshold=3):
    visited = set()
    groups = []

    def bfs(start_point):
        queue = deque([start_point])
        group = []
        
        while queue:
            point = queue.popleft()
            if point in visited:
                continue
            visited.add(point)
            group.append(point)
            
            # Check all other points to see if they are within the threshold distance
            for neighbor in points:
                if neighbor not in visited and manhattan_distance(point, neighbor) <= distance_threshold:
                    queue.append(neighbor)
        return group

    for point in points:
        if point not in visited:
            group = bfs(point)
            groups.append(group)
    
    return groups
```
