```python
input_text = """9 players; last marble is worth 25 points
10 players; last marble is worth 1618 points
13 players; last marble is worth 7999 points
17 players; last marble is worth 1104 points
21 players; last marble is worth 6111 points
30 players; last marble is worth 5807 points
465 players; last marble is worth 71940 points"""
input_lines = input_text.splitlines()
```

```python
import re
from collections import deque

def parse_input(lines):
    games = []
    pattern_input=re.compile('\d+')
    for line in lines:
        player_count, marble_value = pattern_input.findall(line)
        games.append((int(player_count), int(marble_value)))
    return games

def play_game(game):
    scores = {}
    player_count, marble_value = game
    players = deque([player for player in range(player_count)])
    marbles = deque([marble for marble in range(marble_value + 1)])
    # First, the marble numbered 0 is placed in the circle.
    circle = deque([marbles.popleft()])
    #print(f"players: {players}")
    #print(f"marbles: {marbles}")
    #print(f"circle: {circle}")
    while marbles:
        # Then, each Elf takes a turn placing the lowest-numbered remaining marble 
        current_player = players[0]
        # advance players for next pass
        players.rotate(-1)
        #print(f"players: {players}")
        current_marble = marbles.popleft()
        # However, if the marble that is about to be placed has a number which is a multiple of 23
        if current_marble > 0 and current_marble % 23 == 0:
            # something entirely different happens.
            # First, the current player keeps the marble they would have placed, adding it to their score.
            # init score for current_player if needed
            if current_player not in scores:
                scores[current_player] = 0
            scores[current_player] += current_marble
            # In addition, the marble 7 marbles counter-clockwise from the current marble is removed from the circle
            circle.rotate(7)
            #print(circle)
            current_marble = circle.popleft()
            # and also added to the current player's score. The marble located immediately clockwise of the marble
            # that was removed becomes the new current marble.
            scores[current_player] += current_marble
            continue
        # into the circle between the marbles that are 1 and 2 marbles clockwise of the current marble.
        # (When the circle is large enough, this means that there is one marble between the marble that
        # was just placed and the current marble.) The marble that was just placed then becomes the current marble.
        circle.rotate(-2)
        circle.appendleft(current_marble)
        #print(circle)
    return scores

for game in parse_input(input_lines):
    player_count, marble_value = game
    scores = play_game(game)
    #print(scores)
    print(f"{player_count} players; last marble is worth {marble_value} points: high score is {max(scores.values())}")
```
