```python
import re
pattern_input = re.compile(r'Disc #(\d+) has (\d+) positions; at time=(\d+), it is at position (\d+).')
```

```python
def parse_input(input_string):
    """
    Function to parse input
    """
    # get p
    match = pattern_input.match(input_string)
    if match:
        # add to disks
        disks.append(tuple(int(data) for data in match.groups()))
    else:
        # just in case
        print(f"Error matching input string {input_string}")

def machine_state(target_time):
    """
    Function to calculate machine state at a particular time
    I likely won't use this for the solution, rather a modification of it
    to increment time, but useful now to play with the data structure
    """
    state = []
    for disk in disks:
        # get disk definition
        # note, after reviewing the input, start time is always 0, 
        # so we don't need to calculate initial position
        # (id, num_positions, start_time, start_position)
        id, num_positions, start_time, current_position = disk
        # walk time to target_time
        current_time = start_time
        for current_time in range(start_time+1,target_time+1):
            current_position += 1
            # if we are past the end, go to 0 
            if current_position == num_positions:
                current_position = 0
            
        state.append((id, num_positions, current_time, current_position))
    return state
            
def visualize(state, capsule_pos):
    max_size = 0
    for disk in disks:
        if disk[1] > max_size:
            max_size = disk[1]
    current_time = state[0][2]
    print(f"t={current_time}:")
    for idx, disk in enumerate(state):
        chars = list('_'*max_size)
        chars[disk[3]] = ' '
        if capsule_pos == idx + 1:
            if chars[0] == ' ':
                chars[0] = 'o'
            else:
                chars[0] = '!' 
        print(''.join(chars) + "\n")
        
def drop_capsule(drop_time):
    max_size = 0
    for disk in disks:
        if disk[1] > max_size:
            max_size = disk[1]
    capsule_pos = 0
    for my_time in range(drop_time,drop_time+len(disks)+2):
        my_state = machine_state(my_time)
        #visualize(my_state, capsule_pos)
        for idx, disk in enumerate(my_state):
            chars = list('_'*max_size)
            chars[disk[3]] = ' '
            if capsule_pos == idx + 1:
                if chars[0] == ' ':
                    chars[0] = 'o'
                else:
                    return False
        capsule_pos += 1
    return True
        
```

```python
import time

class DropIterator:
    def __init__(self, disk, delay):
        self.disk = disk 
        self.delay = delay
        self.first = disk[1] - disk[3]
        self.current = self.first - self.delay
        while self.current < 0:
            self.current += self.disk[1]
        self.drops = set([self.current])

    def __iter__(self):
        return self

    def __next__(self):
        self.current += self.disk[1]
        self.drops.add(self.current)
        return self.current

def check_results(disk_sets):
    common = disk_sets[disks[0]].drops
    for disk in disks:
        common = common.intersection(disk_sets[disk].drops)
    return common
    
start = time.time()
delay=0
drops = {}
pointer = 0
# initialize iterators
for disk in disks:
    delay+=1
    drops[disk] = DropIterator(disk, delay)

sentinel = 0
while not check_results(drops):
    sentinel += 1
    if sentinel > 100000:
        print("Breaking Loop")
        break
    for disk in disks:
        next(drops[disk])
end=time.time()
print(f"Common: {check_results(drops)}")
print(f"took: {end-start} seconds")
```
